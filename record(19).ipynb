```python
#1. Write a Python program to read a file line by line and store it into a list. 
filename = "file.txt"
with open(filename, 'r') as file:
    lines = file.readlines()

print("File content as list:")
print(lines)
```

```python
# 2. Python program to copy odd lines of one file to other 
source_file = "file.txt"
dest_file = "file4.txt"
with open(source_file, 'r') as infile, open(dest_file, 'w') as outfile:
    for i, line in enumerate(infile, start=1):
        if i % 2 != 0:  # Odd line number
            outfile.write(line)
print("Odd lines copied successfully from file.txt to file4.txt")
```

```python
#3. Write a Python program to read each row from a given csv file and print a list of strings
import csv

filename = "StdData.csv"

with open(filename, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
# 4. Write a Python program to read specific columns of a given CSV file and print the content of the columns. 
import csv
filename = "StdData.csv"
with open(filename, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row['Std Names'], "-", row['Std Total Marks'])
```

```python
 #5. Write a Python program to write a Python dictionary to a csv file. 
# After writing the CSV file read the CSV file and display the content.
import csv
students = [
    {"Name": "Alice", "Age": 22, "City": "Chennai"},
    {"Name": "Bob", "Age": 25, "City": "Delhi"},
    {"Name": "Carol", "Age": 23, "City": "Mumbai"}
]
filename = "students.csv"
# Writing dictionary to CSV
with open(filename, 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["Name", "Age", "City"])
    writer.writeheader()
    writer.writerows(students)
print("Data written successfully to students.csv")
# Reading the same file
with open(filename, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
