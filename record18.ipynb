```python
"""1. Create a class Rectangle with private attributes length and width. Overload ‘<’ operator to
compare the area of 2 rectangles. """
class Rectangle:
    def __init__(self, length, width):   # Correct constructor
        self.__length = length           # Private attributes
        self.__width = width

    def area(self):
        return self.__length * self.__width

    def __lt__(self, other):# Correct operator overload
       return self.area() < other.area()
l=int(input("enetr the length of first rectangle:"))
w=int(input("enetr the width of first rectangle:"))
l2=int(input("enetr the length of 2nd rectangle:"))
w2=int(input("enetr the width of 2nd rectangle:"))
r1 = Rectangle(l,w)
r2 = Rectangle(l2,w2)
if (r1<r2):
    print("rectangle 1 is bigger")
else:
    print("rectangle 2 is bigger")  
```

```python
"""2. Create a class Time with private attributes hour, minute and second. Overload ‘+’ operator to
find sum of 2 time."""
class time:
    def __init__(self,hour,minute,second):
        self.hour=hour
        self.minute=minute
        self.second=second
    def __add__(self,other):
        return (self.hour+other.hour,self.minute+other.minute,self.second+other.second)

t1=time(int(input("Enter the hour of the 1st time ")),
        int(input("eneter the minute :")),
        int(input("Enter the second:")))
t2=time(int(input("Enter the hour of the 2nd time ")),
        int(input("eneter the minute of 2nd:")),
        int(input("Enter the second of 2nd:")))
print("sum of to times",t1+t2)
        
```
